```python
def extract_episodes(txt): 
    check = False
    data = ""
    for i in txt: 
        if i == ")":
           check = False
           return data
        if check == True: 
            data = data + i
        if i == "(":
            check = True
```

```python
df["Title"].apply(extract_episodes)
```

```python
df["Episodes"] = df["Title"].apply(extract_episodes)
```

```python
def extract_time(txt): 
    check = False
    data = ""
    for i in range(len(txt)): 
        if txt[i] == ")":
            for j in range(i+2, i+23): 
                data += txt[j]
            return data
```
